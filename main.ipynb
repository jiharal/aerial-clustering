```python
from src.aerial_clustering.kmeans import KMeansClustering
from src.aerial_clustering.utils import read_json_data

data = read_json_data()
if data is None:
    exit()

# Proses data
clustering = KMeansClustering(max_distance=1000)  # Radius dalam meter
results = clustering.process_aerial_data(data)

print(f"total data {len(results)}")

# Tampilkan hasil
print("\nHasil Pengelompokan Data:")
for i, cluster in enumerate(results, 1):
    print(f"\nKelompok {i}:")
    print(f"Latitude: {cluster['latitude']}")
    print(f"Longitude: {cluster['longitude']}")
    print(f"Timestamp: {cluster['timestamp']}")
    print(f"Jumlah titik dalam radius: {cluster['point_count']}")
```
